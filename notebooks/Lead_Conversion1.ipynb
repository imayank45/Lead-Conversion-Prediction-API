```python
from sklearn.preprocessing import LabelEncoder

# Sample categorical columns
CATEGORICAL_COLUMNS = ['Gender', 'Location', 'LeadSource', 'LeadStatus', 'DeviceType', 'ReferralSource', 'PaymentHistory']

# Load training data (assuming train is your DataFrame)
label_encoders = {}

for col in CATEGORICAL_COLUMNS:
    le = LabelEncoder()
    train[col] = le.fit_transform(train[col])  # Transform data
    label_encoders[col] = le  # Save the encoder
```

```python
train.head()
```

```python
x = train.drop(columns=['LeadID', 'Conversion (Target)'])
y = train['Conversion (Target)']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
rf = RandomForestClassifier(
    n_estimators=100,
    max_depth=5,
    class_weight='balanced',
    random_state=42
)
```

```python
rf.fit(x_train, y_train)
```

```python
y_pred = rf.predict(x_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

```python
# save model into pickle file
import pickle

with open('model_rf2.pkl', 'wb') as file:
    pickle.dump(rf, file)
```
